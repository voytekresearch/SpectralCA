```python
import matplotlib.pyplot as plt
import h5py
from SpectralCV import ecog_pipe as ep
#load data from h5 
h5_file = '../Voytek/scv.h5'
```

```python
with h5py.File(h5_file, 'a') as h5:
    print([k for k in h5['monkey'].items()])
```

```python
def plotMonkey(monkey_path):
#fig, axes = plt.subplots(nrows=4, ncols=1, sharex=True, sharey=True)
# plotting
    with h5py.File(h5_file, 'r') as h5:
        if ("_1") in monkey_path:
            xaxis = np.arange(0,501)
            plt.figure(2)
        elif ("_05") in monkey_path:
            xaxis = np.arange(0,501,2)
            plt.figure(1)
        elif ("_2") in monkey_path:
            xaxis = np.arange(0,500.5,.5)
            plt.figure(3)
        bread = h5[monkey_path]
        #print(len(bread[0][1]))
        #print(chibiBread.shape)
        #xaxis = np.arange(0,501)
        for i in range(5):
            plt.subplot(i+431)
            plt.loglog(xaxis, bread[i][:][:].T)
            plt.title('Condition:' + str(i) + " " + monkey_path[monkey_path.find('/'):])
        plt.subplots_adjust(left=.1, bottom=.1, right=2.3, top=3,
                wspace=.1, hspace=.5)
```

```python
def plotMonkeyAllfs(monkey_mainPath):
    plotMonkey(monkey_mainPath+"_05")
    plotMonkey(monkey_mainPath+"_1")
    plotMonkey(monkey_mainPath+"_2")
```
